```python
import numpy as np
import pandas as pd
from scipy.spatial import  Delaunay
from scipy.spatial.distance import euclidean
from scipy.linalg import solve
from scipy.linalg import lstsq
import matplotlib.pyplot as plt
```

```python
potential_int = pd.read_csv('Data/potential_i.csv', header=None).values.T
x_int = pd.read_csv('Data/pts_i.csv', header=None).values.T
```

```python
points = np.array(x_int)
values = np.array(potential_int)
```

```python
tri = Delaunay(points)
```

```python
diff = [[] for row in range(len(points))] #make diff list
#diff
```

```python
diff_diff = [[] for row in range(len(points))] #make diff list
```

```python
n_vec = [[] for row in range(len(points))]
```

```python
for simplex in tri.simplices:
  #print(simplex)
  point1 = simplex[0]
  point2 = simplex[1]
  point3 = simplex[2] #point of triangle which is index of array
  l1 = euclidean(points[point1], points[point2])
  l2 = euclidean(points[point1], points[point3])
  l3 = euclidean(points[point2], points[point3]) #distance of edge
  diff12 = (values[point2] - values[point1])/l1
  diff13 = (values[point3] - values[point1])/l2
  diff21 = (values[point1] - values[point2])/l1
  diff23 = (values[point3] - values[point2])/l3
  diff31 = (values[point1] - values[point3])/l2
  diff32 = (values[point2] - values[point3])/l3 #calcuate the finite difference
  n12 = points[point2] - points[point1]
  n13 = points[point3] - points[point1]
  n21 = points[point1] - points[point2]
  n23 = points[point3] - points[point2]
  n31 = points[point1] - points[point3]
  n32 = points[point2] - points[point3] #calcuate the vector
  diff[point1].append(diff12)
  diff[point1].append(diff13)
  diff[point2].append(diff21)
  diff[point2].append(diff23)
  diff[point3].append(diff31)
  diff[point3].append(diff32)
  n_vec[point1].append(n12)
  n_vec[point1].append(n13)
  n_vec[point2].append(n21)
  n_vec[point2].append(n23)
  n_vec[point3].append(n31)
  n_vec[point3].append(n32)
```

```python
output = [[] for row in range(len(points))]
```

```python
for i in range(len(points)):
  output[i], residuals, rank, s = lstsq(n_vec[i],diff[i])
```

```python
output = np.reshape(output, (len(points),-1))
```
